#### Status of LA's Vision Zero

To understand the state of car accident data in Los Angeles, we will be examining the Statewide Integrated Traffic Records System (SWITRS) database, maintained by the California Highway Patrol. This data is from 2019, and provides accident rates per city in Los Angeles County. For the purposes of this exercise, we have filtered out the category "Los Angeles" and are looking for the cities that have the highest traffic incident rates. Then we can see if this data aligns with the most dangerous intersections as identified by Vision Zero.

#### Importing data

In order to work with data in python, we need to use libraries that let us handle data exploration. Therefore, we will import the pandas library, and Matplotlib, which helps with charts. 

```python
import pandas as pd
```

```python
# import Matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load the data file
df = pd.read_csv('data/SWITRS_Accidents.csv')
```

#### Preliminary inspection
A quick look at the size of the data.

```python
df.shape
```

```python
df.head()
```

#### Create Charts
Sort the data and create bar charts to display the top 10 cities with the highest incident rates for vehicle-related deaths and injuries.

```python
# First, we'll make a chart of the top 10 cities with the highest rates of pedestrian fatalities from cars.
df_sorted = df.sort_values(by='Pedestrian Involved Fatal',ascending = False)
```

```python
df_sorted[['City','Pedestrian Involved Fatal']].head(10)
```

```python
# plot it
df_sorted.head(10).plot.bar(x='City',
                            y='Pedestrian Involved Fatal',
                            title='Cities with Highest Pedestrian Deaths',
                            color = 'red')
```

```python
df_sorted_ped_injury = df.sort_values(by='Pedestrian Involved Injury',ascending = False)
```

```python
df_sorted_ped_injury[['City','Pedestrian Involved Injury']].head(10)
```

```python
# plot it
df_sorted_ped_injury.head(10).plot.bar(x='City',
                            y='Pedestrian Involved Injury',
                            title='Cities with Highest Pedestrian Injury Rates',
                            color = 'orange')
```

```python
df_sorted_crashes_fatal = df.sort_values(by='Total Fatal Crashes',ascending = False)
```

```python
df_sorted_crashes_fatal[['City','Total Fatal Crashes']].head(10)
```

```python
# plot it
df_sorted_crashes_fatal.head(10).plot.bar(x='City',
                            y='Total Fatal Crashes',
                            title='Cities with Highest Traffic Fatalities',
                            color = 'red')
```

```python
df_sorted_crashes_injury = df.sort_values(by='Total Injury Crashes',ascending = False)
```

```python
df_sorted_crashes_injury[['City','Total Injury Crashes']].head(10)
```

```python
# plot it
df_sorted_crashes_injury.head(10).plot.bar(x='City',
                            y='Total Injury Crashes',
                            title='Cities with Highest Traffic Injuries',
                            color = 'orange')
```
