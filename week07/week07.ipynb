#### Traffic Accidents in Los Angeles County
Exploration of data from Statewide Integrated Traffic Records data (SWITRS), showing traffic accidents broken down by injury or fatality, and by collision type (car-to-car, car-to-pedestrian, car-to-bycycle). This data serves as a baseline for LA's Vision Zero initiative, which was implemented in 2015 and set to run through 2025. The goal of Vision Zero is to reduce traffic injuries and fatalities down to zero. 

This notebook will refine the data based on midterm feedback, and add missing elements to the charts. 

```python
import pandas as pd
# import matplotlib
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline
import numpy as np
```

Next, we import the data from SWITRS

```python
# load the year over year data file
df = pd.read_csv('data/LAFatalities.csv')
```

#### Comparing two groups together
Since I have two categories of data that I want to compare year over year, I will create a grouped bar chart. By not specifying a Y, pandas takes all other columns besides X for the chart.

```python
# make a bar graph of traffic fatalities and injuries
df.head(10).plot.bar(x='Year',
                     title='Traffic Fatalities in LA County 2010 to 2019',
                     color='red')
```

That's great, however I want to show fatalities and injuries in a more distinctive way. Let's try this with a stacked bar plot.

```python
# make a stacked bar plot
df.plot.bar(x='Year',stacked=True,rot=0);
```

That's great! Now let's clean it up with colors, legends, and labels.

```python
df.plot.bar(x='Year',
            stacked=True,
            rot=0,
            color=[(240/255,83/255,101/255), (250/255,188/255,42/255)],             
            figsize=(10,4),
            title='LA County Traffic Fatalities and Injuries 2010 to 2019',
            xlabel='Year',
            ylabel='Persons',
            fontsize='large');
    
```

For comparision, let's try this with a grouped bar chart too, giving us a choice of which view to use.

```python
# Make a grouped bar chart with labels and legends. 
df.plot.bar(x='Year',
            rot=0,
            color=[(240/255,83/255,101/255), (250/255,188/255,42/255)],             
            figsize=(10,4),
            title='LA County Traffic Fatalities and Injuries 2010 to 2019',
            xlabel='Year',
            ylabel='Persons',
            fontsize='large');
```

While the number of traffic fatalities is under 1,000 per year, it is still significant. Traffic injuries impact such a large number of people in LA county. Taken together, this shows the impact that car accidents are having on LA County residents every year, and the numbers seemed to increase from 2015 onward, which is when LA adopted Vision Zero. Clearly Vision Zero is needed, as LA County residents are still getting injured and killed by cars throughout 2015-2019.
